```python
import json

import torch
import torch.nn as nn
from collections import OrderedDict

from MiDaS.midas.midas_net import MidasNet
from YOLOv3.nets.yolo_layers import YOLOLayers
```

```python
# YOLO + MIDAS

class Yolo_Midas(nn.Module):

  def __init__(self, midas_weights_path, yolo_weights_path, yolo_cfg):

    super(Yolo_Midas, self).__init__()

    ## MIDAS PART
    self.midas = MidasNet(midas_weights_path)

    ## RESET/EFFICIENT NET HEAD --> MIDAS BACKBONE  
    self.midas_head = self.midas.pretrained

    ## SCRATCH 
    self.scratch = self.midas.scratch

    # ========================================================== #

    ## YOLO
    # with open(yolo_cfg, 'r') as f:
    #   config = json.load(f)
    self.yolov3 = YOLOLayers(config= d)

    ## READ AND LOAD WEIGHTS
    state_dict = torch.load(yolo_weights_path, map_location = torch.device('cuda' if torch.cuda.is_available() else 'cpu'))
    self.yolov3.load_state_dict(state_dict, strict = False)

    
  def forward(self, X):

    ## HEAD / BACKBONE
    X1 = self.midas_head.layer1(X)
    X2 = self.midas_head.layer2(X1)
    X3 = self.midas_head.layer3(X2)
    X4 = self.midas_head.layer4(X3)

    # YOLO OUTPUT
    yolo_out = self.yolov3.forward(X2, X3, X4)

    ## CONNECTER LAYERS
    X1_rn = self.scratch.layer1_rn(X1)
    X2_rn = self.scratch.layer2_rn(X2)
    X3_rn = self.scratch.layer3_rn(X3)
    X4_rn = self.scratch.layer4_rn(X4)


    ## DECODER MIDAS - ADD/FUSION (add parallel feature maps back to decoding process)
    path_4 = self.scratch.refinenet4(X4_rn)
    path_3 = self.scratch.refinenet3(path_4, X3_rn)
    path_2 = self.scratch.refinenet2(path_3, X2_rn)
    path_1 = self.scratch.refinenet1(path_2, X1_rn)

    # MIDAS OUTPUT
    midas_out = self.scratch.output_conv(path_1)

    return yolo_out, torch.squeeze(midas_out, dim=1)
```

```python
midas_weights_path = '/content/EVA-5-Capstone/EVA5-Capstone-main/MiDaS/model-f6b98070.pt'
yolo_weights_path = '/content/EVA-5-Capstone/EVA5-Capstone-main/YOLOv3/weights/darknet53_weights_pytorch.pth'
yolo_cfg = '/content/EVA-5-Capstone/EVA5-Capstone-main/cfg.json'

model = Yolo_Midas(midas_weights_path, yolo_weights_path, yolo_cfg).to()
```
